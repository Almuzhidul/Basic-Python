#**Import Library**

Terdapat dua library yang akan digunakan dalam membuat dan memanipulasi data tabel yaitu ```Numpy``` dan ```Pandas``` 
*   ```NumPy``` adalah pustaka Python yang digunakan untuk bekerja dengan array.
Ini juga memiliki fungsi untuk bekerja dalam domain aljabar linier, transformasi fourier, dan matriks.
*   ```Pandas``` adalah package open source Python yang paling banyak digunakan untuk ilmu data/analisis data dan tugas machine learning. Pandas dibangun di atas package lain bernama Numpy, yang menyediakan dukungan untuk pengolahan array multi-dimensi 

````python
import numpy as np
import pandas as pd
````

###**Membuat Satu List Menjadi Dataframe**

Membuat list dengan label Nama

````python
Nama = ['Ghifari', 'Nathan', 'Mujhid', 'Alfa', 'Maul', 'Aldi', 'Toriq', 'Yasir', 'Yekti', 'Gina', 'Feni', 'Rima', 'Albays', 'Hafid', 'Ikhsan']
#Memanggil list Nama yang telah di definisikan
Nama
````

Mengubah list sebelumnya menjadi dataframe dengan menggunakan library pandas

````python
data1 = pd.DataFrame(Nama)
data1
````

###**Cara 1. Membuat Dataframe Dari Beberapa List**

Berikut contoh list yang dibuat untuk Umur, Prodi, dan Asal

````python
Umur = [22, 22, 24, 19, 22, 23, 21, 24, 21, 19, 22, 20, 21, 21, 21]
Prodi = ['Sastra Arab', 'Sastra Arab', 'Matematika', 'Teknik Informatika', 'Bimbingan Konseling', 'Teknik Sipil', 'Sistem Informasi', 'Profesi PG Paud', 'PGSD', 'Sastra Indonesia', 'PGSD', 'Teknik Kimia', 'Bimbingan Konseling', 'PGSD', 'Pendidikan Bahasa Inggris']
Asal = ['Jawa Barat', 'Jawa Barat', 'Sulawesi Tenggara', 'Jawa Timur', 'Jawa Timur', 'Jawa Tengah', 'Jawa Barat', 'Jawa Tengah', 'Jakarta', 'Jawa Barat', 'Kalimantan', 'Jawa Timur', 'Jawa Tengah', 'Kalimantan', 'Jawa Barat']
````

List yang telah dibuat kemudian dapat dibuat menjadi dataframe dengan menggunakan zip pada list yang telah di definiskan, kemudian columns diisi dengan nama variabel/feature yang diinginkan

````python
data2 = pd.DataFrame(list(zip(Nama, Umur, Prodi, Asal)), columns =['Nama', 'Umur', 'Program_Studi', 'Asal_Provinsi'])
data2
````

###**Cara 3. Membuat Dataframe Dari Beberapa List**
Cara ke 3 yaitu kita dapat membuat dataframe dengan menggunakan dictionary

````python
Nama = ['Ghifari', 'Nathan', 'Mujhid', 'Alfa', 'Maul', 'Aldi', 'Toriq', 'Yasir', 'Yekti', 'Gina', 'Feni', 'Rima', 'Albays', 'Hafid', 'Ikhsan']
Umur = [22, 22, 24, 19, 22, 23, 21, 24, 21, 19, 22, 20, 21, 21, 21]
Prodi = ['Sastra Arab', 'Sastra Arab', 'Matematika', 'Teknik Informatika', 'Bimbingan Konseling', 'Teknik Sipil', 'Sistem Informasi', 'Profesi PG Paud', 'PGSD', 'Sastra Indonesia', 'PGSD', 'Teknik Kimia', 'Bimbingan Konseling', 'PGSD', 'Pendidikan Bahasa Inggris']
Asal = ['Jawa Barat', 'Jawa Barat', 'Sulawesi Tenggara', 'Jawa Timur', 'Jawa Timur', 'Jawa Tengah', 'Jawa Barat', 'Jawa Tengah', 'Jakarta', 'Jawa Barat', 'Kalimantan', 'Jawa Timur', 'Jawa Tengah', 'Kalimantan', 'Jawa Barat']
# Membuat Dictionary
dict = {'Nama': Nama, 'Umur': Umur, 'Program Studi': Prodi, 'Asal Provinsi': Asal }  
data4 = pd.DataFrame(dict) 
data4
````
